```python
import matplotlib.pyplot as plt
import numpy as np
```

# 2D plotting

As follows is shown a basic plot using the famous libraries **Numpy** and **Matplotlib**.

```python
x_data = np.linspace(0, 2*np.pi, 150)
y_data = np.sin(x_data)
print(len(y_data))

fig, ax = plt.subplots()
ax.plot(x_data,y_data)
ax.set_title('Function sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
plt.show() #Not necessary line of code because jupyter detects our intentions and calls the method automatically at the end of the cell
```

# 3D plotting

Now to increase the complexity it's now plotted a 3D graph of the following multivariable function:
$$f(x,y) = x + y + 2 $$

## Plotting points

```python
ax = plt.axes(projection='3d') #Creates an instance of an object from the plt library that allows us to plot on it
ax.scatter(1,3,5)
ax.set_title('(1,3,5)')#Sets the title of the graph
plt.show()
```

## Plotting lines

```python
ax = plt.axes(projection='3d')
x_data = np.linspace(0,50,200)
y_data = np.linspace(0,50,200)
z_data = np.sin(x_data)*np.cos(y_data)

ax.plot(x_data,y_data,z_data)
ax.set_title('F(x,y) = sin(x)*cos(y)')
plt.show()
```

## Plotting 3D surfaces

```python
ax = plt.axes(projection='3d')

x_data = np.arange(-5,5,0.1) #Creates an array of values between 0 and 50 with a 0.1 step
y_data = np.arange(-5,5,0.1)

X, Y = np.meshgrid(x_data,y_data) #Creates two multidimensional arrays to build all combinations of x and y values
Z = np.sin(X) + np.cos(Y)
ax.plot_surface(X,Y,Z,cmap='plasma') #Cmap creates a color map according to the options provided in the documentation
ax.set_title('F(x,y) = sin(x) + cos(y)')
plt.show()
```
